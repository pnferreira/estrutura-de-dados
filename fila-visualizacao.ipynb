```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches

class Fila:
    def __init__(self):
        self.queue = []

    def enfileirar(self, dataval):
        self.queue.append(dataval)

    def desenfileirar(self):
        if len(self.queue) <= 0:
            return ("Nenhum elemento na fila")
        else:
            self.queue.pop(0)

    def plot_fila(self):
        plt.clf()  # Limpa a figura antes de desenhar a nova fila.
        fig, ax = plt.subplots()
        ax.set_ylim([0, 1])  # Define o limite do eixo y.
        ax.set_xlim([0, max(12, len(self.queue) + 2)])  # Define o limite do eixo x.
        ax.axis('off')  # Remove os eixos.
        width = 1  # Define a largura fixa.
        gap = 0.5  # Define a lacuna entre a bancada e os elementos da fila.
        
        # Adiciona cada retângulo (elemento da fila) ao gráfico.
        for i, item in enumerate(self.queue):
            color = 'lightcoral' if i == 0 else 'lightblue'
            rect = patches.Rectangle((len(self.queue) - 1 - i, 0.25), width, 0.5, linewidth=1, edgecolor='black', facecolor=color)
            ax.add_patch(rect)
            plt.text(len(self.queue) - 0.5 - i, 0.5, item, ha='center', va='center', fontsize=10, rotation='vertical')
        
        # Adiciona a "bancada" à frente da primeira pessoa.
        bancada = patches.Rectangle((len(self.queue) + gap, 0.25), 2, 0.2, linewidth=1, edgecolor='black', facecolor='saddlebrown')
        ax.add_patch(bancada)
        
        plt.show()

fila = Fila()
```

```python
fila.enfileirar(5)
```

```python
fila.plot_fila()
```
