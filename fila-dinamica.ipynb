# Fila Dinâmica

Em python, podemos implementar a fila dinâmica utilizando:

1. listas
2. classe nativa do python para filas (mais eficiente)

```python
class Fila:
    def __init__(self):
        self.fila = []

    def enfileirar(self, elemento):
        self.fila.append(elemento)
        
    def desenfileirar(self):
        if not self.is_empty():
            elemento = self.fila.pop(0)
            return elemento
        
    def size(self):
        return len(self.fila)
        
    def primeiro(self):
        return self.fila[0]
    
    def is_empty(self):
        return len(self.fila) == 0

```

```python
fila = Fila()
fila.fila
fila.enfileirar(1)
fila.enfileirar(2)
fila.fila
```

```python
fila.desenfileirar()
fila.fila
```
