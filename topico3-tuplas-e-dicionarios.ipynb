<div>

1. Declaração de dicionários
2. Manipulação de dicionários
    1. Operações básicas
    2. Iterando pelas chaves e valores

```python
d = {"nome":"Poliana", "idade":24}
print(d)

print(d["nome"])

d["nome"] = "Poliana Ferreira"
print(d)

d["profissao"] = "professora"
print(d)

del d["idade"]
print(d)
```

```python
for chave in d.keys():
    print(chave)
    
for valores in d.values():
    print(valores)
    
for c, v in d.items():
    print(c, v)
```
